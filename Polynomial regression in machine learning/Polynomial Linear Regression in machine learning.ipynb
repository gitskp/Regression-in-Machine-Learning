### Polynomial Linear Regression

```python
# Data Preprocessing Template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values




```

```python
# Fitting the Regression Model to the dataset
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
# Fitting the Ploynomial Regression Model to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X) # here plynomial of X is created
lin_reg2= LinearRegression() # now fit the polynomial of X into linera regression model
lin_reg2.fit(X_poly,y)
```

```python
# visulazing the linear regression 
plt.scatter(X,y,color= 'g',label="green color are actual data points ") # it si the actual data points
plt.plot(X,reg.predict(X),color = 'blue',label="blue color are the predicted data points") # it is the predicted data points
plt.xlabel("Position label")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# visulazing the polynomial linear regression 
plt.scatter(X,y,color= 'g',label="green color are actual data points ") # it si the actual data points
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color = 'blue',label="blue color are the predicted data points") # it is the predicted data points
plt.xlabel("Position label")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# predicting the linear regression model
reg.predict(6.5)
```

```python
# predicting the ploynomial linear regression model
lin_reg2.predict(poly_reg.fit_transform(6.5))
```
