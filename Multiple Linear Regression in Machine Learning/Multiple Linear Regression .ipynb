```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Multiple Linear Regression

dataset = pd.read_csv('50_Startups.csv')
dataset.head()
```

```python
# remove the Profit column because it is depenedent bariavle y
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values


# Encoding categorical data
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 3] = labelencoder_X.fit_transform(X[:, 3])
onehotencoder = OneHotEncoder(categorical_features = [3])
X = onehotencoder.fit_transform(X).toarray()


# avoid the dummy column trap
X = X[:,1:]



```

```python
# split the dataset into training set and testing set

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,random_state=0)
```

```python
# fitting the Multile linear regression to the training set

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
# predicte  the  test result
y_pred = reg.predict(X_test)
y_pred
```

```python
# Building the optimal model  using Backward Elimination

import statsmodels.formula.api as sm
X = np.append(arr= np.ones((50,1)).astype(int),values = X,axis = 1)
X_opt = X[:,[0,1,2,3,4,5]] #step1
reg_OLS= sm.OLS(endog =y,exog = X_opt).fit() #step 2
reg_OLS.summary() # step 3
# if P>|t| then remove it #step 4




    
```

```python
X_opt = X[:,[0,1,2,3,5]] #step1
reg_OLS= sm.OLS(endog =y,exog = X_opt).fit() #step 2
reg_OLS.summary() #step 3
```
